## LINE OF REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
np.random.seed(42)
data = {
    'INDICATOR': np.random.choice(['Indicator_A', 'Indicator_B', 'Indicator_C'], 100),
    'UNIT': np.random.choice(['Unit_1', 'Unit_2', 'Unit_3'], 100),
    'UNIT_NUM': np.random.choice([1, 2, 3], 100),
    'STUB_NAME': np.random.choice(['Stub_A', 'Stub_B', 'Stub_C'], 100),
    'STUB_NAME_NUM': np.random.choice([1, 2, 3], 100),
    'STUB_LABEL': np.random.choice(['Label_A', 'Label_B', 'Label_C'], 100),
    'STUB_LABEL_NUM': np.random.choice([1, 2, 3], 100),
    'YEAR': np.random.choice([2018, 2019, 2020, 2021, 2022], 100),
    'YEAR_NUM': np.random.choice([2018, 2019, 2020, 2021, 2022], 100),
    'AGE': np.random.choice(['0-18', '19-35', '36-50', '51+'], 100),
    'AGE_NUM': np.random.choice([1, 2, 3, 4], 100),
    'ESTIMATE': np.random.rand(100) * 100,
    'FLAG': np.random.choice(['Flag_A', 'Flag_B', 'Flag_C'], 100)
}
A = pd.DataFrame(data)
missing_values = A.isnull().sum()
A_cleaned = A.dropna(subset=['ESTIMATE'])
X_cleaned = A_cleaned['YEAR_NUM'].values.reshape(-1, 1)
y_cleaned = A_cleaned['ESTIMATE'].values
model_cleaned = LinearRegression()
model_cleaned.fit(X_cleaned, y_cleaned)
y_pred_cleaned = model_cleaned.predict(X_cleaned)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YEAR_NUM', y='ESTIMATE', data=A_cleaned, label='Actual data', edgecolor=None)
plt.plot(A_cleaned['YEAR_NUM'], y_pred_cleaned, color='red', label='Regression line')
plt.title('Linear Regression: ESTIMATE vs. YEAR_NUM (Cleaned Data)')
plt.xlabel('Year')
plt.ylabel('Estimate')
plt.legend()
plt.show()
```
